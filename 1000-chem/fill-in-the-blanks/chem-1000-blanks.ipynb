```python
path = "fill-in-the-blanks.txt"
text_file = open(path, "r", encoding="utf8")
text = text_file.read()
questions_text, answers_text= text.split("الإجابة")
questions = {}
answers = {}
```

##### questions

```python
# Remove page numbers
questions_text = questions_text.replace("293", "")
questions_text = questions_text.replace("294", "")
questions_text = questions_text.replace("295", "")
questions_text = questions_text.replace("296", "")
questions_text = questions_text.replace("297", "")
questions_text = questions_text.replace("298", "")
questions_text = questions_text.replace("299", "")
```

```python
# Replace blanks
pattern = r"\.{5,15}"
add_newline= lambda match : " ..... "
questions_text = re.sub(pattern, add_newline, questions_text)
questions_text
```

```python
# fix non aligning questions
mis_aligned_text = ".....  أكثر الموازين شيوعا في معامل الكيمياء -877، بينما الأحماض المعدنية يدخل فى تركيبها  .....  الأحماض العضوية تستخلص من -878 ..... عناصر  .....  يذوب السكر فى الماء عن طريق تكوين روابط -879"
questions_text = questions_text.replace(mis_aligned_text, "")
questions[877] = "أكثر الموازين شيوعا في معامل الكيمياء ....."
questions[878] = "الأحماض العضوية تستخلص من .....، بينما الأحماض المعدنية يدخل في تركيبها ....."
questions[879] = "يذزب السكر في الماء عن طريق تكوين روابط ....."
```

```python
# fix parentheses
import re
pattern = re.compile(r'\)\s*([a-zA-Z]{1,3})\s*\(')
def fix_parentheses(match):
   m = match.group(0).replace(")", "").replace("(", "").strip()
   return f" ({m}) "
questions_text = re.sub(pattern, fix_parentheses, questions_text)
```

```python
# parse questions
pattern = r"([89]\d{2}-|1\d{3}-)"
add_newline= lambda match : "\n" + match.group(0)
questions_text_sep = re.sub(pattern, add_newline, questions_text)
for line in questions_text_sep.split("\n")[1:]:
  number, text = line.split("-")
  questions[int(number)] = text.strip()
```

```python
# remove questions having complex chars
q_ids = set([893, 936, 938, 939, 945, 958, 965, 968, 972])
# remove questions with answers having complex chars
a_ids = set([902, 914, 925, 936, 958, 968, 972, 982, 988, 991])
for id in q_ids.union(a_ids):
  _ = questions.pop(id)
```

```python
# remove end point
pattern = re.compile(r'(?<!\.)(\.|\.\.)(?=[^\.]*$)')
for k,v in questions.items():
  questions[k] = pattern.sub('', v)
```

```python
# difficult parentheses
questions[943] = "حجم 2 mol من غاز الهيدروجين ..... حجم 1 mol من غاز ثانى أكسيد الكبريـت في (STP)"
```

```python
# Drop hard fill in the blank (4 blanks)
questions.pop(895)
questions.pop(918)
questions.pop(984)
```

##### answers

```python
# Remove page numbers
answers_text = answers_text.replace("300", "")
answers_text = answers_text.replace("301", "")
```

```python
# parse answer
pattern = r"([689]\d{2}-|10\d{2}-)"
add_newline= lambda match : "\n" + match.group(0)
answers_text_sep = re.sub(pattern, add_newline, answers_text)
passed_887 = 0 # To fix a double number mistake in original text
passed_818 = 0 # To fix a wrong number mistake in original text
for line in answers_text_sep.strip().split("\n")[1:]:
  number, text = line.split("-")
  number = int(number)
  if answers.get(number): # pass for the second time to number 887
    passed_887 = 1
  if number == 818: # pass for a wrong number
    number = 897
    passed_818 = 1
  if number == 603:
    number = 903
  number = number + passed_887 + passed_818
  answers[number] = text.strip()
```

```python
# remove answers with questions having complex chars
q_ids = set([893, 936, 938, 939, 945, 958, 965, 968, 972])
# remove answers having complex chars
a_ids = set([902, 914, 925, 936, 958, 968, 972, 982, 988, 991])
for id in q_ids.union(a_ids):
  _ = answers.pop(id)
```

```python
# Fix seperated answers
answers[885] = "النسبة بين المساحة الكلية إلى الحجم"
answers[918] = "التسخين، إضافة الكحول، إضافة الحماض المعدنية المركزة، التعرض للإشعاع"
```

##### combine

```python
from collections import defaultdict
groups = defaultdict(list)
for k,v in answers.items():
  items = v.split("،")
  groups[len(items)].append(k)
```

```python
import random
all_questions = []
for k,v in questions.items():
  question = {}
  question["question"] = v
  question["answer"] = {}
  answer = answers[k]
  correct_option = random.randint(0,3)
  random_ids = random.sample([s for s in groups[len(answer.split("،"))] if s!=k], 4)
  for i in [0, 1, 2, 3]:
    if i==correct_option:
      question["answer"][f"option_{correct_option}"] = answer
    else:
      question["answer"][f"option_{i}"] = answers[random_ids[i]]
  question["answer"]["correct_option"] = correct_option
  all_questions.append(question)
```
