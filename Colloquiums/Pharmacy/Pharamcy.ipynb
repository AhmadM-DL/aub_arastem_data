```python
import json, os
```

```python
root = "/content/drive/MyDrive/AUB Research/Arabic Dataset/Colloquiums/Pharmacy"
raw = os.path.join(root, "raw")
files = []
for f in os.listdir(raw):
  if not "KEY" in f:
    files.append(f)
print(files)
```

```python
files_2 = {}
for f in files:
  key = f.split(".")[0] + "_KEY.json"
  data = json.load(open(os.path.join(raw, f), "r"))
  key_data = json.load(open(os.path.join(raw, key), "r"))
  key_data = key_data[list(key_data.keys())[0]]
  if "correct_answer" in key_data[0]:
    key_data = {item["no"]:item["correct_answer"] for item in key_data}
  elif "correct answer" in key_data[0]:
    key_data = {item["no"]:item["correct answer"] for item in key_data}
  else:
    raise Exception("Key not found")
  data_2 = []
  for q in data:
    q_number = q["no"]
    q["correct_option"] = key_data[int(q_number)]
    data_2.append(q)
  files_2[f] = data_2
```

```python
files_3 = {}
for f, data in files_2.items():
  data = { int(v["no"]):v for v in data}
  indices = sorted(data.keys())
  data_3 = []
  for i in indices:
    data_3.append(data[i])
  files_3[f] = data_3
```

```python
files_4 = {}
for f, data in files_3.items():
  for q in data:
    q["answer"] = {}
    for i,choice in enumerate(q["choices"]):
      q["answer"][f"option_{i}"] = choice
    mapping = {'A':0, 'B':1, 'C':2, 'D':3, 'E':4, 'F':5}
    q["answer"]["correct_option"] = mapping[q["correct_option"]]
    del q["choices"]
    del q["correct_option"]
  files_4[f] = data
```

```python
processed = os.path.join(root, "processed")
for f, data in files_4.items():
  json.dump(data, open(os.path.join(processed, "P"+f), "w"), ensure_ascii=False, indent=4)
```
