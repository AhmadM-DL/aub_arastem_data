```python
# 1000_chem
chem_1000_mcq = json.load(open("./1000-chem/choose-correct-answer/1000-chem-mcq-qa.json", "r", encoding="utf-8"))
chem_1000_define_term = json.load(open("./1000-chem/define-the-term/1000-chem-define-qa.json", "r", encoding="utf-8"))
chem_1000_true_false = json.load(open("./1000-chem/true-false/1000-chem-true-false-qa.json", "r", encoding="utf-8"))
chem_1000_blanks = json.load(open("./1000-chem/fill-in-the-blanks/1000-chem-blanks.json", "r", encoding="utf-8"))

# 1500_bio
bio_1500_define = json.load(open("./1500-bio/define-the-term/1500-bio-define-qa.json", "r", encoding="utf-8"))
bio_1500_mcq = json.load(open("./1500-bio/choose-the-correct-answer/1500-bio-mcq-qa.json", "r", encoding="utf-8"))

# bedaya
bedaya = json.load(open("./Beadaya/beadaya_qa.json", "r", encoding="utf-8"))

# alloschool
alloschool = json.load(open("./AllSchool/allschool_qa.json", "r", encoding="utf-8"))

# kiyas
kiyas_biology = json.load(open("./Kiyas/biology/kiyas_biology_mcq.json", "r", encoding="utf-8"))
kiyas_chemistry = json.load(open("./Kiyas/chemistry/kiyas_chemistry_mcq.json", "r", encoding="utf-8"))
kiyas_math = json.load(open("./Kiyas/math/kiyas_math_mcq.json", "r", encoding="utf-8"))
kiyas_physics = json.load(open("./Kiyas/physics/kiyas_physics_mcq.json", "r", encoding="utf-8"))

# manara
manara_physics = json.load(open("./Manara/manara_physics_mcq.json", "r", encoding="utf-8"))

# awael
awael_information_technology = json.load(open("./Awael/awael_computer_mcq.json", "r", encoding="utf-8"))

# ksu
ksu_information_technology = json.load(open("./KSU/ksu_it_mcq.json", "r", encoding="utf-8"))

# Kau
kau_information_technology = json.load(open("./KAU/kau_it_mcq.json", "r", encoding="utf-8"))

# egypt
egypt_information_technology = json.load(open("./Egypt/egypt_it_mcq.json", "r", encoding="utf-8"))

# colloqiums
dentistry = json.load(open("./Colloquiums/Dentistry/dentistry_qa.json", "r", encoding="utf-8"))
medicine = json.load(open("./Colloquiums/Medicine/medicine_qa.json", "r", encoding="utf-8"))
pharmacy = json.load(open("./Colloquiums/Pharmacy/pharmacy_qa.json", "r", encoding="utf-8"))
```

```python
def get_columnar(qa, level, subject, resource):
  columnar = {}
  columnar["question"] = qa["question"]
  columnar["option_0"] = qa["answer"]["option_0"]
  columnar["option_1"] = qa["answer"]["option_1"]
  columnar["option_2"] = qa["answer"]["option_2"] if "option_2" in qa["answer"] else ""
  columnar["option_3"] = qa["answer"]["option_3"] if "option_3" in qa["answer"] else ""
  columnar["option_4"] = qa["answer"]["option_4"] if "option_4" in qa["answer"] else ""
  columnar["correct_option"] = qa["answer"]["correct_option"]
  columnar["level"] = level
  columnar["subject"] = subject
  columnar["resource"] = resource
  return columnar
```

```python
AMCQ__columnar = []

# Chem
for qa in chem_1000_define_term:
    AMCQ__columnar.append(get_columnar(qa, "collage", "chemistry", "book(https://platform.almanhal.com/Details/Book/53455)"))
for qa in chem_1000_mcq:
    AMCQ__columnar.append(get_columnar(qa, "collage", "chemistry", "book(https://platform.almanhal.com/Details/Book/53455)"))
for qa in chem_1000_blanks:
    AMCQ__columnar.append(get_columnar(qa, "collage", "chemistry", "book(https://platform.almanhal.com/Details/Book/53455)"))
for qa in chem_1000_true_false:
    AMCQ__columnar.append(get_columnar(qa, "collage", "chemistry", "book(https://platform.almanhal.com/Details/Book/53455)"))

# Bio
for qa in bio_1500_define:
    AMCQ__columnar.append(get_columnar(qa, "collage", "biology", "book(https://platform.almanhal.com/Details/Book/63242)"))
for qa in bio_1500_mcq:
    AMCQ__columnar.append(get_columnar(qa, "collage", "biology", "book(https://platform.almanhal.com/Details/Book/63242)"))

# Bedaya
for qa in bedaya:
    level = "primary" if qa["grade"] in ["1", "2", "3", "4", "5", "6", "7", "8", "9"] else "secondary"
    AMCQ__columnar.append(get_columnar(qa, level, qa["subject"], f"web({qa['resource']})"))

# Alloschool
for qa in alloschool:
    level = "secondary"
    subject = "science" if qa["subject"] == "physics-chemistry" else qa["subject"]
    AMCQ__columnar.append(get_columnar(qa, level, subject, f"web({qa['resource']})"))

# Manara
for qa in manara_physics:
    AMCQ__columnar.append(get_columnar(qa, "collage", "physics", "web(https://manara.edu.sy/downloads/files/1682768884_Questions.pdf)"))

# Kiyas
for qa in kiyas_biology:
    AMCQ__columnar.append(get_columnar(qa, "secondary", "biology", "web(https://eqiyas.com/)"))
for qa in kiyas_chemistry:
    AMCQ__columnar.append(get_columnar(qa, "secondary", "chemistry", "web(https://eqiyas.com/)"))
for qa in kiyas_math:
    AMCQ__columnar.append(get_columnar(qa, "secondary", "math", "web(https://eqiyas.com/)"))
for qa in kiyas_physics:
    AMCQ__columnar.append(get_columnar(qa, "secondary", "physics", "web(https://eqiyas.com/)"))

# awael
for qa in awael_information_technology:
    AMCQ__columnar.append(get_columnar(qa, "collage", "information technology", "web(https://www.awa2el.net/ar/content/%D9%86%D9%85%D8%A7%D8%B0%D8%AC-%D8%A7%D9%85%D8%AA%D8%AD%D8%A7%D9%86%D8%A7%D8%AA-%D8%A7%D9%84%D9%85%D8%B3%D8%AA%D9%88%D9%89-%D9%81%D9%8A-%D8%A7%D9%84%D8%AC%D8%A7%D9%85%D8%B9%D8%A7%D8%AA)"))

# ksu
for qa in ksu_information_technology:
    AMCQ__columnar.append(get_columnar(qa, "collage", "information technology", "web(https://faculty.ksu.edu.sa/sites/default/files/syl_nzry_khtbr_nhyy_41.pdf)"))

# kau
for qa in kau_information_technology:
    AMCQ__columnar.append(get_columnar(qa, "collage", "information technology", "web(https://www.kau.edu.sa/Content-0010403-AR-34970)"))

# egypt
for qa in egypt_information_technology:
    AMCQ__columnar.append(get_columnar(qa, "collage", "information technology", "web(https://www.slideshare.net/slideshow/ss-42742711/42742711)"))

# colloquim
for qa in dentistry:
  AMCQ__columnar.append(get_columnar(qa, "collage", "dentistry", f"web({qa['resource']})"))
for qa in medicine:
  AMCQ__columnar.append(get_columnar(qa, "collage", "medicine", f"web({qa['resource']})"))
for qa in pharmacy:
  AMCQ__columnar.append(get_columnar(qa, "collage", "pharmacy", f"web({qa['resource']})"))
```

```python
import pandas as pd
import re
AMCQ_df = pd.DataFrame(AMCQ__columnar)

# Remove trailing ":"
AMCQ_df["question"] = AMCQ_df["question"].apply(lambda s: re.sub(r':$', '', s))
# Remove trailing "?"
AMCQ_df["question"] = AMCQ_df["question"].str.replace("؟", "")
# remove final dot .
AMCQ_df["question"] = AMCQ_df["question"].str.replace(r'(?<!\.)\.$', '', regex=True)
# unify dots ....
AMCQ_df["question"] = AMCQ_df["question"].apply(lambda s: re.sub(r"\.{4,}", ".....", s))
# Remove elongations
AMCQ_df["question"] = AMCQ_df["question"].apply(lambda s: re.sub(r"ـ", "", s))
AMCQ_df["option_0"] = AMCQ_df["option_0"].apply(lambda s: re.sub(r"ـ", "", s))
AMCQ_df["option_1"] = AMCQ_df["option_1"].apply(lambda s: re.sub(r"ـ", "", s))
AMCQ_df["option_2"] = AMCQ_df["option_2"].apply(lambda s: re.sub(r"ـ", "", s))
AMCQ_df["option_3"] = AMCQ_df["option_3"].apply(lambda s: re.sub(r"ـ", "", s))
AMCQ_df["option_4"] = AMCQ_df["option_4"].apply(lambda s: re.sub(r"ـ", "", s))
# Remove wrong paranthesis
AMCQ_df["question"] = AMCQ_df["question"].apply(lambda s: re.sub(r'\)(.*?)\(', r'(\1)',s))
AMCQ_df["option_0"] = AMCQ_df["option_0"].apply(lambda s: re.sub(r'\)(.*?)\(', r'(\1)',s))
AMCQ_df["option_1"] = AMCQ_df["option_1"].apply(lambda s: re.sub(r'\)(.*?)\(', r'(\1)', s))
AMCQ_df["option_2"] = AMCQ_df["option_2"].apply(lambda s: re.sub(r'\)(.*?)\(', r'(\1)', s))
AMCQ_df["option_3"] = AMCQ_df["option_3"].apply(lambda s: re.sub(r'\)(.*?)\(', r'(\1)', s))
AMCQ_df["option_4"] = AMCQ_df["option_4"].apply(lambda s: re.sub(r'\)(.*?)\(', r'(\1)', s))
```
